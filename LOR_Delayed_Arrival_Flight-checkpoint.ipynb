## LAS Vegas Delay in Arrival Flight Using Logistic Regression
The model aims to build a **Logistic Regression model** to predict a **Delay in Arrival of Flights**. The dataset undergoes comprehensive preprocessing, including handling missing values, encoding categorical variables. The model performance is evaluated using metrics such as **accuracy_score** and **classification_report** .

## Define column transformer to scale and encode

```python
df['carrier_code'] =df['carrier_code'].astype(str)
```

```python
categorical_cols = ['carrier_code']
encode = OneHotEncoder(drop = 'first',sparse_output=False)
encoded_array = encode.fit_transform(df[categorical_cols])
encoded_columns = encode.get_feature_names_out(categorical_cols)
df_encoded = pd.DataFrame(encoded_array,columns=encoded_columns)
df = df.drop(columns=categorical_cols).reset_index(drop=True)
df=pd.concat([df,df_encoded],axis=1)
```

```python
numerical_cols = ['scheduled_elapsed_time_minutes','actual_elapsed_time_minutes','taxi-in_time_minutes','delay_carrier_minutes','delay_weather_minutes','delay_national_aviation_system_minutes','delay_security_minutes','delay_late_aircraft_arrival_minutes']
scaler = StandardScaler()
scaled_cols = pd.DataFrame(scaler.fit_transform(df[numerical_cols]),columns=numerical_cols)
```

### Droped columns not needed

```python
df=df.drop(columns =['date','flight_number','tail_number','origin_airport','scheduled_arrival_time','actual_arrival_time','wheels-on_time']) 
```

### Define features (X) and target (y)

```python
x = df.drop(columns =['delay_late_aircraft_arrival_minutes'])
y = df['delay_late_aircraft_arrival_minutes']
```

```python
x
```

## Step 4: Model Building

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

### Training the model

```python
LOR = LogisticRegression()
```

```python
LOR
```

```python
LOR.fit(x_train,y_train)
```

```python
y_pred=LOR.predict(x_test)
```

```python
acc=accuracy_score(y_test,y_pred)
acc
```

```python
classification=classification_report(y_test,y_pred)
print(classification)
```

```python
r2=r2_score(y_test,y_pred)
r2
```

```python
mae = mean_absolute_error(y_test,y_pred)
mae
```

```python
mse= mean_squared_error(y_test,y_pred)
mse
```

```python
rmse = np.sqrt(mean_squared_error(y_test,y_pred))
rmse
```

## Tunning

```python
ridge_model = Ridge(alpha=1.0)
ridge_score = cross_val_score(ridge_model, x_train, y_train, cv=5)
```

```python
lasso_model = Lasso(alpha=1.0)
lasso_score = cross_val_score(lasso_model, x_train, y_train, cv=5)
```

```python
ElasticNet_model = ElasticNet(alpha=1.0)
ElasticNet_score = cross_val_score(ElasticNet_model , x_train, y_train, cv=5)
```

```python
ridge_model.fit(x_train, y_train)
ridge_predictions = ridge_model.predict(x_test)
ridge_r2 = r2_score(y_test, ridge_predictions)
ridge_r2*100
```

```python
lasso_model.fit(x_train, y_train)
lasso_predictions = lasso_model.predict(x_test)
lasso_r2 = r2_score(y_test, lasso_predictions)
lasso_r2*100
```

```python
ElasticNet_model.fit(x_train, y_train)
ElasticNet_predictions = ElasticNet_model.predict(x_test)
ElasticNet_r2 = r2_score(y_test, ElasticNet_predictions)
ElasticNet_r2*100
```
